```python
# Upload and plot multiple data files in Google Colab

import matplotlib.pyplot as plt
from google.colab import files
import io

# Upload files
uploaded = files.upload()

# Create figure
plt.figure(figsize=(10, 6))

# Read and plot each file
for filename, filedata in uploaded.items():

    wavelength = []
    intensity = []

    # Read file line by line
    content = io.StringIO(filedata.decode('utf-8'))

    for line in content:
        parts = line.strip().split()

        # Skip invalid lines
        if len(parts) < 2:
            continue

        try:
            x = float(parts[0])
            y = float(parts[1])

            wavelength.append(x)
            intensity.append(y)

        except ValueError:
            # Skip header or bad lines
            continue

    # Plot this dataset
    plt.plot(wavelength, intensity, label=filename)

# Labels and styling
plt.xlabel("Wavelength")
plt.ylabel("Intensity")
plt.title("Uploaded Spectra")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
